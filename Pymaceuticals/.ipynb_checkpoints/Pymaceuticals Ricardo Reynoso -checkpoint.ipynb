```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os
# __________________________________________________________________________________
    # File to Load (Remember to Change These)
mouse_drug_data_to_load = "../Pymaceuticals/mouse.csv"
clinical_trial_data_to_load = "../Pymaceuticals/clinical.csv"
# __________________________________________________________________________________
    # Read the Mouse and Drug Data and the Clinical Trial Data
    # mousedf stands for the mouse data in csv
    # clinicaldf stands for the clinical data in csv
mousedf =pd.read_csv(mouse_drug_data_to_load)
clinicaldf =pd.read_csv(clinical_trial_data_to_load)
# __________________________________________________________________________________
    # Sort both Data Frames by 'Mouse ID', ascending=True
mousedf.sort_values(by=['Mouse ID'], ascending=True)
clinicaldf.sort_values(by=['Mouse ID'], ascending=True)
# __________________________________________________________________________________
# Combine the data into a single dataset
    # Merge both data base into a single one, merge1
    # Rename Columns for easier Handling
    # Align Data                           
merge1 = pd.merge(mousedf, clinicaldf, how='left', on='Mouse ID')
merge1 = merge1.rename(index=str, columns={'Mouse ID': 'Mouse','Timepoint':'TP','Tumor Volume (mm3)':'Tumor Vol','Metastatic Sites':'Metastastic' })
#__________________________________________________________________
    # Display the data table for preview, sort Data
merge1.sort_values('Mouse', ascending=True)
merge1.head()
```

## Display the last values in the Data Frame by Mouse 


```python
merge1.tail()
```

## Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 


```python
# Convert to DataFrame

    #Define meanvolumedf as the Data Frame by grouping First instance: Drug, TP then cross with Tumor Vol
    #Apply unstack to display whole Treatments
    #Rename Tumor Vol as Average Volume
meanvolumedf = pd.DataFrame(merge1.groupby(['Drug', 'TP']).mean()['Tumor Vol'])
meanvolumedf.rename(columns={'Tumor Vol':'Tumor Average Volume (mm3)'},inplace=True)
meanvolumedf.head()
```

## Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint


```python
# Convert to DataFrame
    
    #Call SEM function to return unbiased standard error of the mean over requested axis, it's an estimated value
    #Get into Pandas to Get DataFrame, unstack to display Treatments
    #Rename Tumor Vol as SEM ERROR for Tittle
    #Define semvoldf as the DataFrame

semvoldf = pd.DataFrame(merge1.groupby(['Drug', 'TP']).sem()['Tumor Vol'])
semvoldf.rename(columns={'Tumor Vol':'SEM Vol'},inplace=True)
#______________________________________________________________________________________________________________________
# Preview DataFrame
semvoldf.head()
```

## Minor Data Munging to Re-Format the Data Frames


* Average Volume per Treatment

```python
    #Re Format the Data Frame, unstack the Rows in Columns (0) and remove the Row Tumor Vol
meanvolumedf = meanvolumedf.unstack(level = 0)
meanvolumedf.columns = meanvolumedf.columns.get_level_values(1)
meanvolumedf
```

* SEM Volume

```python
semvoldf = semvoldf.unstack(level = 0)
semvoldf.columns = semvoldf.columns.get_level_values(1)
semvoldf
```

* Generate the Plot (with Error Bars)

   1.- PLOT WITH THE WHOLE DRUGS
  * DAYS Vs TUMOR VOLUME BY TREATMENT


```python
plt.figure(figsize = (12,7))
plt.title('TUMOR RESPONSE OVER TIME Vs. TREATMENT')
plt.xlabel('NUMBER OF DAYS')
plt.ylabel('TUMOR VOLUME SIZE= (mm3)')
plt.xticks(np.arange(0, meanvolumedf.index.max()+2, 10))
plt.xlim(0, meanvolumedf.index.max() + 2) 

    #FROM MATPLOB LIBRARY AND MATPLOT.ORG

    # independent variable
    
    # Markers and colors as requested for the 4 treatments, the remaining at my choice.
x_axis = meanvolumedf.index
list_markers = ['o','v', '^', 's', 'p','d','h','+','P','x']
list_colors=['r','c','b','g','m','k','y','lime','turquoise','orange']
count = 0

    #plot each tumor volume column with standard error
for c in meanvolumedf.columns:
    plt.errorbar(x_axis, meanvolumedf[c], semvoldf[c], marker=list_markers[count],color=list_colors[count],linestyle = '--',label=str(c))
    count += 1
#LEGEND AND FORMAT
plt.grid()
plt.tight_layout()
plt.legend(frameon=True, edgecolor='Black')
plt.savefig('../Pymaceuticals/SEM_PLOT1.png')
plt.show()
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
#Get the Mouse Survival rate by counting the Mouse and grouping by Drug Treatment 
#against Time Point, those who still live, mousesurv

# Convert to DataFrame
mousesurv = merge1.groupby(['Drug', 'TP']).count()['Mouse']
mousesurv = mousesurv.unstack(level = 0)
# Preview DataFrame
mousesurv
```

```python
# Generate the Plot (Accounting for percentages)
#plot options
plt.figure(figsize = (12,7))
plt.title('Percentage Survival Rate Mouse')
plt.xlabel('Days')
plt.ylabel('Survival Rate (%)')
plt.xlim(0, mousesurv.index.max())

# Set the Suvival per Treatment throughout the Dayrs(Time Point) as the x_axis
x_axis = mousesurv.index 

   #Define the Markers and Colors as requested       
list_markers = ['o','v', '^', 's', 'p','d','h','+','P','x']
list_colors=['r','c','b','g','m','k','y','lime','turquoise','orange']
          
    # Iterate the values of the Survival divided by he Standard Error per Treatment(Drug) throughout the Days (Time Point)       
count = 0 
    #GEt the % for Survival Rate per Treatment, Divided by Data Frame by row and incrementing the Columns 
    #Treatment value stated in the TimePoint (days)
          
for c in mousesurv:
    y_axis = mousesurv[c]/mousesurv.loc[0,c] * 100 
    plt.plot(x_axis, y_axis, linestyle = '--', marker=list_markers[count],color=list_colors[count],label=str(c))
    count += 1
    
#LEGEND AND FORMAT
plt.grid()
plt.tight_layout()
plt.legend(frameon=True, edgecolor='Black')
plt.savefig('../Pymaceuticals/SURVIVAL_PLOT.png')
plt.show()

```

## Summary Bar Graph

```python
    # Get the Percentage in Delta (Changes) for the Tumor Volumen by Treatment (Drug)
    # Set Deltatumor

    # Extract the data from the DataFrame Average of the Tumor Volume (meanspread)
    # To obtain the increment is required to read from the last Row and its corresponding TumorVolume by Treatment(Drug)
    # minus the first value and sotre it to plot the values. The .loc function helps by going to the row 45 and read all
    # the values from the Column (treatment) and substract it and dividing by the first Value (Initial), by rows and columns

Deltatumor = ((meanvolumedf.loc[45,:] - meanvolumedf.loc[0,:])/meanvolumedf.loc[0,:] )*100
Deltatumor
```

* Tumor Change over 45 Days FOR THE WHOLE TREATMENTS

```python
#plot options
plt.figure(figsize = (12,7))
plt.title('Tumor Volume Change over 45 Day Treatment')
plt.ylabel('Survival Rate (%)')
xlabels = Deltatumor.index
#puts drug names as labels
plt.xticks(np.arange(len(xlabels)), xlabels)
plt.axhline(y=0, color = 'black') 


#graphs % change of tumor as bar graph
plt.bar(np.arange(10), 
        Deltatumor, 
        # colors each bar appropriately based on + or - values by using a conditional
        color = ['red' if Deltatumor[r] > 0 else 'green' for r in np.arange(len(xlabels))])

#count for x coordinate location of value label
count = 0

#loops through tumor change data and places the value in the correct position based on + or - values
for r in Deltatumor:
    if r < 0:
        y_coor = -3.5
    else:
        y_coor = 2
    plt.text(count, y_coor, str(round(r, 1)) + '%', ha = 'center', color = 'white')
    count += 1
plt.grid()
plt.tight_layout()  
plt.savefig("../Pymaceuticals/TUMOR45.png")
plt.show()
```
